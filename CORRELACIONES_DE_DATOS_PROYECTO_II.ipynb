**Pruebas de Normalidad**
Se prueba normalidad en el conjunto de datos con distribución normal generado aleatoriamente. Podemos probar si los valores de las variables tienen una distribución normal.

```python
import numpy as np
data_puntos = np.random.normal (0, 1, 100)
data_puntos
```

Creamos el histograma con los datos generados aleatoriamente:

```python
plt.hist(data_puntos, edgecolor='red', linewidth=1)
```

Creamos el gráfico Quantile-Quantile para corroborar la normalidad de los datos:




```python
import pylab
import scipy.stats as stats

stats.probplot(data_puntos, dist='norm', plot=pylab)
pylab.show()
```

Aplicamos Shapiro Wilks para probar normalidad: Cuando se tiene un P_Value > 0,05 significa que los valores tienen una distribución normal.

---



```python
from scipy.stats import shapiro
estadistico, p_value= shapiro(data_puntos)
print('estadistico=% 3f, p_value=% 3f' % (estadistico, p_value))
```

**Calcular Correlación y su Significancia**
Se obtiene la correlación de todos los pares de variables. Debido a que no todas las variables presentan una distribución normal se utiliza el método de **spearman**. Alli se observan las correlaciones de todas las variables, se evidencian variables con una correlación cercana a cero, es decir que no están correlacionadas, hay unas que tienen una correlación moderada, con valores positivos y negativos y tambien se prsentan variables con correlaciones altas.

```python
datos_correlación=datos.corr(method='spearman')
datos_correlación
```

**Se crea el mapa de calor de las correlaciones:** Esto sirve para identificar las variables altamente correlacionada ya sea de manera negativa o positiva. El grafico se interpreta de la siguiente manera: los valores de las correlaciones se identifican en la barra del lado derecho, las de color azul fuerte o rojo intenso son las mas fuertes, las azules son correlaciones negativas y las rojas son positivas.

```python
sns.heatmap(datos_correlación,
            xticklabels=datos_correlación.columns,
            yticklabels=datos_correlación.columns,
            cmap='coolwarm'
            )
```
